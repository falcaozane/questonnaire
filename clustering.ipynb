```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn import preprocessing 
  


# Load the data
data = pd.read_csv('synthetic_leadership_data.csv')
data.shape
data.head(10)


```

```python
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])

# Define features
X = data.drop('leadership_style', axis=1)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method graph
import matplotlib.pyplot as plt
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Choose the optimal number of clusters (e.g., 3)
optimal_clusters = 9
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add the cluster labels to the original data
data['cluster'] = clusters

# Evaluate the clustering using Silhouette Score
silhouette_avg = silhouette_score(X_scaled, clusters)
print("Silhouette Score:", silhouette_avg)
```

```python
# To visualize the clusters
import seaborn as sns
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis')
plt.title('Clusters of Users')
plt.show()

# Analyze the clusters
cluster_analysis = data.groupby('cluster').mean()
print(cluster_analysis)
```

```python
import pandas as pd

# Load dataset
data = pd.read_csv('profiles.csv')

# One-Hot Encoding the 'leadership_style' column
#data_encoded = pd.get_dummies(data, columns=['leadership_style'], drop_first=True)

# Now `data_encoded` includes the one-hot encoded leadership styles
data.head(10)
```

```python
# Dropping the 'leadership_style' column for clustering
X = data.drop(columns=['leadership_style'])

# Now `X` contains only the Q1-Q26 responses
X.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming you're using the dataset without 'leadership_style'
scaler = StandardScaler()

# Scaling only the numerical features (Q1–Q26)
X_scaled = scaler.fit_transform(X)

# `X_scaled` can now be used for K-Means clustering
```

```python
from sklearn.decomposition import PCA

# Reduce the dataset to 2 components for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# `X_pca` contains the reduced 2D data
```

```python
from sklearn.cluster import KMeans

# Apply KMeans clustering
kmeans = KMeans(n_clusters=9, random_state=42)
clusters = kmeans.fit_predict(X_scaled)  # or X_pca or X_tsne

# Adding the cluster labels to the dataset (optional)
data['cluster'] = clusters
```

```python
import matplotlib.pyplot as plt

# Visualize the clusters using PCA or t-SNE reduced data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')
plt.title('Clusters of Users')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Reduce the dataset to 3 components for 3D visualization
pca = PCA(n_components=3)
X_pca_3d = pca.fit_transform(X_scaled)

# Now X_pca_3d contains 3D data that can be used for visualization
```

```python
from sklearn.cluster import KMeans

# Apply KMeans clustering
kmeans = KMeans(n_clusters=9, random_state=42)
clusters = kmeans.fit_predict(X_scaled)  # or X_pca_3d for clustering on reduced data

# Optional: Add cluster labels to the original dataset
data['cluster'] = clusters
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the clusters
scatter = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], c=clusters, cmap='viridis')

# Add labels and a color bar
ax.set_title('3D Clusters of Users')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
fig.colorbar(scatter, label='Cluster')

plt.show()
```
